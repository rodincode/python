```python
import os
from time import ctime
from googlesearch import search
import requests
from bs4 import BeautifulSoup
import random
import bs4 as bs
import urllib.request
import requests
from gtts import gTTS
from playsound import playsound
import subprocess
import pyaudio
import speech_recognition as sr
import pyttsx3
import sys

# Initialize the recognizer
r = sr.Recognizer()

# Function to convert text to
# speech


def SpeakText(command):

    # Initialize the engine
    engine = pyttsx3.init()
    engine.say(command)
    engine.runAndWait()


print("Hello I am your Personalised Chatbot, Don Jr.")
SpeakText("  Hello I am your Personalised Chatbot, Don Jr.")
SpeakText("ask Something:: ")
#query=input("What do you want to know: \n")


def app(query):
    myprogram = query.split()
    myprogram = myprogram[1]
    command = r'C:\Users\dell\AppData\Roaming\Microsoft\Windows\Start Menu\Programs' + \
        str(myprogram)+'.exe'
    try:
        subprocess.Popen(command)
    except FileNotFoundError:
        SpeakText("No Such File Available, Bro!")


def Speech_to_Text():
    # It takes microphone input from the user and returns string output

    r = sr.Recognizer()
    # use the microphone as source for input.
    with sr.Microphone() as source2:

        # wait for a second to let the recognizer
        # adjust the energy threshold based on
        # the surrounding noise level
        r.adjust_for_ambient_noise(source2, duration=0.2)

        # listens for the user's input
        audio2 = r.listen(source2)

    try:
        # Using ggogle to recognize audio
        query = r.recognize_google(audio2)
        query = query.lower()
        print("Did you say:: " + query)

    except Exception as e:
        # print(e)
        SpeakText("I can't understand...")
        return "noinfo"
    return query


query = Speech_to_Text()

if "how are you" in query:
    SpeakText("I am fine")

elif "what time is it" in query:
    SpeakText(ctime())

elif "arvie" in query:
    SpeakText("Arvie is my creator")

elif "jarvis" in query:
    SpeakText("I'm there for you")

elif "coronavirus" in query:
    print("This is a new strain of coronavirus that has not been previously identified in humans. Coronaviruses are a large family of viruses that are known to cause illness in humans and animals. In humans, this large family of viruses are known to cause illness ranging from the common cold to Severe Acute Respiratory Syndrome (SARS)."
    )
    SpeakText(
        "This is a new strain of coronavirus that has not been previously identified in humans. Coronaviruses are a large family of viruses that are known to cause illness in humans and animals. In humans, this large family of viruses are known to cause illness ranging from the common cold to Severe Acute Respiratory Syndrome (SARS)."
    )

elif "where is" in query:
    data = query.split(" ")
    location = data[2]
    SpeakText("Hold on, I will show you where " + location + " is.")
    os.system("chromium-browser https://www.google.nl/maps/place/" + location +
              "/&amp;")

elif "vikas" in query:
    print(
        "Ekdum se inhone Waqt badal diya, Jazbaad badal diye, Gaadi palat di")
    SpeakText(
        "Ekdumm inho nee Waq qat bhaa dal diya jaz baad bhaa dal diye gha adhi pall lat dee"
    )
    
elif "open" in query:
    app(query)

else:
    SpeakText("Okay searching for" + query + "on the net")
    query = query.replace(" ", "+")
    # Sushant+Singh+Rajput
    URL = f"https://google.com/search?q={query}"

    USER_AGENT = "Mozilla/5.0 (Macintosh; Intel Mac OS X 10.14; rv:65.0) Gecko/20100101 Firefox/65.0"

    headers = {"user-agent": USER_AGENT}

    resp = requests.get(URL, headers=headers)
    soup = BeautifulSoup(resp.content, "html.parser")
    result = []
    for i in soup.find_all('div', class_='r'):
        anchors = i.find_all('a')
        if anchors:
            link = anchors[0]['href']
            title = i.find('h3').text
            item = {"title": title, "link": link}
            result.append(item)

    # print(result)
    # len(result)
    search = result[0]
    # search=result[random.randint(0,len(result)-1)]
    search = search['link']
    print(str(search))

    x = urllib.request.urlopen(search).read()
    soup = bs.BeautifulSoup(x, "lxml")
    para = []
    x = soup.find('p').getText()
    print(x)
    SpeakText(x)

    for paragraph in soup.find_all("p"):
        print(paragraph.text)
        SpeakText(paragraph.text) #what this is doing
```
